```python
# x and y product summed up
index = 0
x_y_product_sum = 0
for value in x:
    x_y_product_sum = x_y_product_sum + (x[index] * y[index])
    index = index + 1

print(x_y_product_sum)
```

```python
# x summed and y summed and multiplied
x_sum = 0
y_sum = 0
for value in x:
    x_sum = x_sum + value
for value in y:
    y_sum = y_sum + value

x_sum_multiply_y_sum = x_sum * y_sum

print(x_sum)
print(y_sum)
print(x_sum_multiply_y_sum)
```

```python
# Values squared and summed

x2_sum = 0
y2_sum = 0

for value in x:
    x2_sum = value*value + x2_sum
for value in y:
    y2_sum = value*value + y2_sum

print(x2_sum)
print(y2_sum)
```

```python
#check if number of index are the same
if len(x) == len(y):
    n = len(x)
else: 
    print("data does not match")

#similarity coefficient

correlation_coefficient = (n*x_y_product_sum - x_sum*y_sum)/math.sqrt((n*x2_sum - x_sum*x_sum)*(n*y2_sum - y_sum*y_sum))

print(correlation_coefficient)
```
